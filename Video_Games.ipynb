```python
df_NA_Sales= df[["Publisher","NA_Sales"]]
```

```python
df_NA_Sales
```

```python
df_NA = df_NA_Sales.groupby("Publisher").sum()["NA_Sales"].sort_values(ascending=False).head(1)
```

```python
df_EU_Sales= df[["Publisher","EU_Sales"]]
```

```python
df_EU_Sales
```

```python
df_EU= df_EU_Sales.groupby("Publisher").sum()["EU_Sales"].sort_values(ascending=False).head(1)
```

```python
df_JP_Sales= df[["Publisher","JP_Sales"]]
```

```python
df_JP_Sales
```

```python
df_JP= df_JP_Sales.groupby("Publisher").sum()["JP_Sales"].sort_values(ascending=False).head(1)
```

```python
df_Other_Sales = df[["Publisher","Other_Sales"]]
```

```python
df_Other_Sales
```

```python
df_others = df_Other_Sales.groupby("Publisher").sum()["Other_Sales"].sort_values(ascending=False).head(1)
```

```python
[df_NA , df_EU, df_JP, df_others]
```

```python
pd.concat([df_NA , df_EU, df_JP, df_others])
```
